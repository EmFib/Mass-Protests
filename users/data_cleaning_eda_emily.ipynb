```python
def string_to_number(string):
    
#     tuple_of_prefixes = ('Around', 'at least', 'about', 'nearly', 'up to', 'as many as', 'almost' ,'over', 'some')
#     for word in tuple_of_prefixes:
#         if str(string).lower().startswith(word):
#             string = string.strip(word)
#     if str(string).lower().startswith('around'):
#             string = string.strip('around ')


    if str(string)[0].isdigit():
        return re.sub("[^0-9]", "", string)
    
    prefixes = ['Around', 'at least', 'about', 'nearly', 'up to', 'as many as', 'almost' ,'over', 'some']
    if str(string).lower().startswith(tuple(prefixes)):
        string = re.sub("[^0-9]", "", string)
    
#     if any(map(l))
    
    # removes "About"/"about" at start of phrase 
    elif str(string).lower().startswith('about'):
        return str(string).lower().strip('about ')
    

    
    # remove "<" or ">" in front of number 
    elif str(string).startswith('>'):
        return str(string).strip('>')
    elif str(string).startswith('> '):
        return str(string).strip('> ')
    elif str(string).startswith('<'):
        return str(string).strip('<')
    elif str(string).startswith('< '):
        return str(string).strip('< ')
    
        
    elif str(string) == 'thousands':
        return '1000'
    
    # remove "more than" or "less than" in front of number 
    elif str(string).lower().startswith('more than'):
        return str(string).lower().strip('more than')
    elif str(string).lower().startswith('more than '):
        return str(string).lower().strip('more than ')
    elif str(string).lower().startswith('less than'):
        return str(string).lower().strip('less than')
    elif str(string).lower().startswith('less than '):
        return str(string).lower().strip('less than ')
    
    elif ('hundreds of thousands') in str(string).lower():
        return '100000'
    
    elif str(string).lower().startswith('hundreds'):
        return '900'

    
    else:
        return string
```

```python
def extract_50k(item):
    
    if isinstance(item, list):
#         print('List found')
        if len(item) == 0:
            return None 
        l = []
        for i in item:
        
            converted_i = re.sub("[^0-9]", "", str(i))
            l.append(converted_i)
        final_list = []
        for i in l: 
#             print(i)
            if len(i) == 0:
                pass
            elif float(str(i)) >= float(100_000):
#                 print('correct')
                final_list.append(float(str(i))) 
        return final_list
            
            
    else: 
        converted_item = re.sub("[^0-9]", "", str(item))
#         print('no list found')
        if len(converted_item) == 0:
            return None 
        elif float(converted_item) >= 100_000:
            return float(converted_item) 
        else: 
            return None 
```

```python
mass['greater_than_50k'] = mass['numbers_in_notes'].apply(extract_50k)
```
