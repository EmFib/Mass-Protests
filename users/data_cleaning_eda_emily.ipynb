```python
def string_to_number(string):
    
#     tuple_of_prefixes = ('Around', 'at least', 'about', 'nearly', 'up to', 'as many as', 'almost' ,'over', 'some')
#     for word in tuple_of_prefixes:
#         if str(string).lower().startswith(word):
#             string = string.strip(word)
#     if str(string).lower().startswith('around'):
#             string = string.strip('around ')


    if str(string)[0].isdigit():
        return re.sub("[^0-9]", "", string)
    
    if str(string).lower().startswith(('Around', 'at least', 'about', 'nearly', 'up to', 'as many as', 'almost' ,'over', 'some')):
        return re.sub("[^0-9]", "", string)
    
    # removes "About"/"about" at start of phrase 
    elif str(string).lower().startswith('about'):
        return str(string).lower().strip('about ')
    

    
    # remove "<" or ">" in front of number 
    elif str(string).startswith('>'):
        return str(string).strip('>')
    elif str(string).startswith('> '):
        return str(string).strip('> ')
    elif str(string).startswith('<'):
        return str(string).strip('<')
    elif str(string).startswith('< '):
        return str(string).strip('< ')
    
        
    elif str(string) == 'thousands':
        return '1000'
    
    # remove "more than" or "less than" in front of number 
    elif str(string).lower().startswith('more than'):
        return str(string).lower().strip('more than')
    elif str(string).lower().startswith('more than '):
        return str(string).lower().strip('more than ')
    elif str(string).lower().startswith('less than'):
        return str(string).lower().strip('less than')
    elif str(string).lower().startswith('less than '):
        return str(string).lower().strip('less than ')
    else:
        return string
```
