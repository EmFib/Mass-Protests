```python
import pandas as pd 
import matplotlib.pyplot as plt 
import numpy as np 

from sklearn.preprocessing import OneHotEncoder
```

```python
protests = pd.read_csv('./data/mmALL_073120_csv.csv')
```

```python
protests.columns
```

```python
protests.shape
```

```python
responses = protests[['stateresponse1', 'stateresponse2', 'stateresponse3', 'stateresponse4',
       'stateresponse5', 'stateresponse6', 'stateresponse7']]
```

```python
responses.info()
```

```python
responses.isnull().sum()
```

```python
responses.fillna('no response', inplace=True)
```

```python
responses
```

```python
onehot = OneHotEncoder(dtype=np.int, sparse=True)
```

```python
responses = pd.DataFrame(onehot.fit_transform(responses[['stateresponse1', 'stateresponse2', 'stateresponse3', 'stateresponse4',
       'stateresponse5', 'stateresponse6', 'stateresponse7']]).toarray(),
                         columns=onehot.get_feature_names(['stateresponse1', 'stateresponse2', 'stateresponse3', 'stateresponse4',
       'stateresponse5', 'stateresponse6', 'stateresponse7']))
```

```python
onehot.get_feature_names()
```

```python
responses
```

```python
responses.columns
```

```python
responses['arrests'] = responses['stateresponse1_arrests'] + responses['stateresponse2_arrests'] + responses['stateresponse3_arrests'] + responses['stateresponse4_arrests'] + responses['stateresponse5_arrests'] + responses['stateresponse6_arrests'] + responses['stateresponse7_arrests'] 
```

```python
responses
```

```python
responses['accomodation'] = responses['stateresponse1_accomodation'] + responses['stateresponse2_accomodation'] + responses['stateresponse3_accomodation'] + responses['stateresponse4_accomodation'] + responses['stateresponse5_accomodation'] + responses['stateresponse6_accomodation'] + responses['stateresponse7_accomodation'] 
responses['beatings'] = responses['stateresponse1_beatings'] + responses['stateresponse2_beatings'] + responses['stateresponse3_beatings'] + responses['stateresponse4_beatings'] + responses['stateresponse5_beatings'] + responses['stateresponse6_beatings'] + responses['stateresponse7_beatings'] 
responses['crowddispersal'] = responses['stateresponse1_crowd dispersal'] + responses['stateresponse2_crowd dispersal'] + responses['stateresponse3_crowd dispersal'] + responses['stateresponse4_crowd dispersal'] + responses['stateresponse5_crowd dispersal'] + responses['stateresponse6_crowd dispersal'] 
responses['ignore'] = responses['stateresponse1_ignore'] + responses['stateresponse2_ignore'] + responses['stateresponse3_ignore']
responses['killings'] = responses['stateresponse1_killings'] + responses['stateresponse2_killings'] + responses['stateresponse3_killings'] + responses['stateresponse4_killings'] + responses['stateresponse5_killings'] + responses['stateresponse6_killings'] + responses['stateresponse7_killings'] 
responses['no_response'] = responses['stateresponse1_no response'] + responses['stateresponse2_no response'] + responses['stateresponse3_no response'] + responses['stateresponse4_no response'] + responses['stateresponse5_no response'] + responses['stateresponse6_no response'] + responses['stateresponse7_no response'] 
responses['shootings'] = responses['stateresponse1_shootings'] + responses['stateresponse2_shootings'] + responses['stateresponse3_shootings'] + responses['stateresponse4_shootings'] + responses['stateresponse5_shootings'] 
responses
```

```python
responses.columns
```

```python
state_responses = responses[]
```

```python
state_responses = responses[['arrests', 'accomodation', 'beatings',
       'crowddispersal', 'ignore', 'killings', 'no_response', 'shootings']]
state_responses
```

```python
state_responses['no_response'] = state_responses['no_response'].map(lambda b: 1 if b > 0 else 0)
state_responses
```
